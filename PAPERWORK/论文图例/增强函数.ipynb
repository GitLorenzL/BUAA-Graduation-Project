```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = ['Arial Unicode MS']

# 定义 x 的取值范围
x = np.linspace(-1, 1, 100)

# 定义 y = x 的函数
y1 = x

# 定义 y = 5x^5 的函数
y2 = 5 * x ** 5

# 创建一个画布，并在画布上绘制两个函数的图像
fig, ax = plt.subplots()
ax.plot(x, y1, label='原始信号')
ax.plot(x, y2, label='增强信号')

# 设置 x 轴和 y 轴的显示范围
ax.set_xlim(-1, 1)
ax.set_ylim(-5, 5)

# 设置 x 轴和 y 轴的刻度密度为 0.2
ax.set_xticks(np.arange(-1, 1.2, 0.2))
ax.set_yticks(np.arange(-5, 5.2, 1))

# 显示坐标轴
ax.grid()

# 添加图例
ax.legend()

# 添加 x 轴和 y 轴标签
ax.set_xlabel('眼动信号')
ax.set_ylabel('反馈信号')

# 显示图像
plt.show()
```
